```python
#Obtem a pagina sem fazer verificacao do certificado
nUrl = "https://www.bondfaro.com.br/celular-e-smartphone.html?mn=1"
```

```python
#Analisa a pagina obtida
pg = rq.get(nUrl, verify = False)
analisador_pg = bs(pg.content, 'html.parser')
```

```python
#Mostra a pagina do analisador, limitar a 1000 caracteres
print(analisador_pg.prettify()[:1000])
```

```python
#Buscando a PRIMEIRA tag a
analisador_pg.find('a')
```

```python
#Buscando TODAS as tags a
analisador_pg.find_all('a')
```

```python
#Analisar a pagina, ver o que acontece ao pesquisar a classe PRICE
#Traz os dados dos smartphones
dados_celulares = analisador_pg.find_all(class_ = 'price')

for x in dados_celulares:
    print(x.prettify())
```

```python
#Quero saber os PRECOS dos MODELOS de smartphones da MARCA SAMSUNG

#Tag com class='title' -> MODELO
#Tag p span com VALUE -> PREÇO
#Tag p span com SELLERS -> VENDEDORES

smartphones = []
dados = analisador_pg.find_all('li', class_ = re.compile("prd-new"))

for x in dados:
    modelos = x.find('p', class_='title').text
    precos = x.find('span', class_='value').text
    lojas = x.find('span', class_='sellers').text
    smartphones.append(dict(zip(['Modelo', 'Preco', 'Lojas'], [modelos, precos, lojas])))
display(smartphones)
```
